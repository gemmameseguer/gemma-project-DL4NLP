# Data Exploration

```python
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.read_csv('Gender_bias_spanish.csv')
dataset
```

```python
sns.countplot(data=dataset, x="Label")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.boxplot(dataset['Contents'].str.len())
plt.xlabel('Text Length')
plt.ylabel('Length')
plt.title('Boxplot of Text Length')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot(dataset['Contents'].str.split().apply(len))
plt.xlabel('Text Length (Number of Words)')
plt.ylabel('Number of Words')
plt.title('Boxplot of Text Length (Number of Words)')
plt.show()
```
